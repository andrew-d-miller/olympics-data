```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Save csv file names
file_one = "summer_olympics.csv"
file_two = "gdp.csv"
```

```python
#Import csv files
olympicsI_df = pd.read_csv(file_one, encoding="ISO-8859-1")
gdp_df = pd.read_csv(file_two, encoding="ISO-8859-1")
```

```python
#View olympics file
olympicsI_df.head()
```

```python
#Drop NaN
olympicsI_df.dropna()
```

```python
#Checking for file types and number of data points
olympicsI_df.info()
```

```python
#Setting year to be a string
olympicsI_df = olympicsI_df.fillna(0)
olympicsI_df["Year"] = olympicsI_df["Year"].astype(int)
olympicsI_df["Year"] = olympicsI_df["Year"].astype(str)
```

```python
#Viewing gdp df
gdp_df.head()
```

```python
#Fix South Korea
olympicsI_df["Country"].replace({"Korea, South": "South Korea", "Korea, North": "North Korea"}, inplace = True)
```

```python
#Concatenating the country and year for the purpose of merging the two files
olympicsI_df["combo"] = olympicsI_df["Country"] + olympicsI_df["Year"]
olympicsI_df.head()
```

```python
#Counting the number if datapoints from each city
olympicsI_df.City.value_counts()
```

```python
#Creating a function for assigning host countries
def country_getter(city):
    if city == 'Beijing':
        return('China')
    elif city == 'Sydney':
        return('Australia')
    elif city ==  "Athens":
        return("Greece")
    elif city == "Atlanta":
        return("United States")
    elif city == "Barcelona":
        return("Spain")
    elif city == "Montreal":
        return("Canada")
    elif city == "Seoul":
        return("South Korea")
    elif city == "Los Angeles":
        return("United States")
    elif city == "Moscow":
        return("Russia")
    else:
        return('country not found')
```

```python
#Applying host country function
olympicsI_df['host country'] = olympicsI_df.City.apply(country_getter)
```

```python
#Double Checking host country counts
olympicsI_df["host country"].value_counts()
```

```python
#Concatenating gdp country and year for merge purposes
gdp_df["combo"] = gdp_df["Entity"] + gdp_df["Year"].map(str)
```

```python
#Checking for new combo column
gdp_df.head()
```

```python
korea_list = olympicsI_df[olympicsI_df["Country"]== "South Korea"]
korea_list
```

```python
#Merging two dataframes
olympics_df = pd.merge(olympicsI_df, gdp_df, on="combo", how = "left")
```

```python
olympics_df.drop('Code', inplace=True, axis=1)
olympics_df.drop('Country_Code', inplace=True, axis=1)
olympics_df.drop('Entity', inplace=True, axis=1)
olympics_df.rename(columns = {"Year_y": "GDP Year", "Year_x": "Olympics Year"}, inplace = True)
```

```python
from scipy.stats import linregress
```

```python
#Grouping by Country and GDP Year to find medal count 
country_year_df = olympics_df.groupby(["Country", "GDP Year"]).count()
#country_year_df.head()

#Creating new Dataframe 
medal_df = country_year_df[["Medal"]]
medal_df
```

```python
#Grouping by Country and GDP Year to find GDP for each country for each Olympic Year
country_year_df2 = olympics_df.groupby(["Country", "GDP Year"]).max()

#Creating new Dataframe with just GDP 
gdp_df = country_year_df2[["GDP (constant 2010 US$)"]]
gdp_df
```

```python
#Merging two dataframes together
merge_df = pd.merge(medal_df, gdp_df, on =["Country", "GDP Year"])
merge_df
```

```python
#x and y values for our scatter plot
x_values = merge_df["GDP (constant 2010 US$)"]
y_values = merge_df["Medal"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

#Finding values
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,11)) + "x + " + str(round(intercept,2))

#Plotting our data and line
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(8000000000000,50),fontsize=15,color="red")
plt.xlabel("Country GDP")
plt.ylabel("Number of Medals")
plt.title("Country GDP and Number of Medals")

print(f"The r-squared value is: {rvalue**2}")
```
